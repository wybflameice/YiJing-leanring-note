# <center>**紫微斗数学习笔记**</center>

## 前言了解
&emsp;&emsp;中国的五术，即山医卜命相。“山”，指的是修道、修炼、养生，层次最高。“医”，指的是医术，治病的方法。“卜”，指的是八卦、三式（奇门、六壬、太乙）等预测术。“命”，指的是以出生时间研究个人命运的方法，种类繁多，主要有四柱命理学和紫微斗数命理学。“相”，指的是从人、物、地等方面研究命运的方法，包括手面相、风水地理等术。<br>
&emsp;&emsp;紫微斗数是陈抟[tuán]老祖（华山道士）总结出天人相应、星人相应的哲理。紫微斗数因紫微垣居中天而得名，它是通过南北斗紫微星群的运行规律，揭示星象与人之间的相互对应影响，以及通过星象的变化、运行来反应人事的信息运动轨迹，从而由星知人，预测出人一生的命运曲线。紫微斗数内容十分丰富，信息容量巨大，在中国五大神数中名列榜首，号称“天下第一神数”。紫微斗数不必取用神，排出命盘后，根椐各宫的星情及其三方四正星曜的扶抑情况，即可进行论断，易学易懂，比较直观、具体、细致。但因紫微斗数长期深藏于高层社会中，虽古老而新奇，经验尚欠老成，很多观点还有争议，且易学而难精，有待不断开发利用、实践总结。四柱预测须取准用神，用神一错，全盘皆非，但相对来讲，四柱命理流行较广，理论经验较丰富。所以，最好将紫微和四柱两者结合起来，互相取长补短，这样信息内容就比较全面准确。<br>
