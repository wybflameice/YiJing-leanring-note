# <center>**太上感应篇学习笔记**</center>

《太上感应篇》是道家经典，被誉为“古今第一善书”。阐述了因果报应理论，劝人遵守道德规范，时刻止恶修善。参考：[太上感应篇原文](http://www.taishangganyingpian.com/)<br>

**太上曰，祸福无门，惟人自召，善恶之报，如影随形。**<br>
*太上老君说，灾祸和福气没有固定的门路，不是天生注定谁就倒霉，这些都是人自己招来的，自身的念头、行为、选择，一步步把福祸引到身边。善与恶的报应，就像影子跟随身体一样，走到哪，跟到哪儿。*<br>

**是以天地有司过之神，依人所犯轻重，以夺人算。算减则贫耗，多逢忧患。人皆恶之，刑祸随之，吉庆避之，恶星灾之，算尽则死。**<br>
*所以说，天地之间有专门记录人过错的神明，会根据一个人所犯错误的轻重，来削减他的寿命。寿算一旦减少，就会变得贫穷衰败，还会经常遭遇忧愁和祸患。人们也都会厌恶他，刑罚与灾祸会随之而来；吉祥喜庆都避开他，凶星灾厄会降临到他身上。等到寿算被削减完就会死去。*<br>

**又有三台北斗神君，在人头上，录人罪恶，夺其纪算。又有三尸神，在人身中，每到庚申日，辄上诣天曹，言人罪过，月晦之日，灶神亦然。**<br>
*还有三台星和北斗的神君，在人的头顶之上，奏报人们的罪过，并削减相应的年寿（一纪12年，一算100天）。又有“三尸神”住在人的身体里面，每到庚申日（六十甲子日循环中的57天），就会上到天庭的官署，奏报人们的罪过。到了阴历每月最后一天（晦日），灶神也会这样做。*<br>

**凡人有过，大则夺纪，小则夺算。其过大小，有数百事，欲求长生者，先须避之。**<br>
*凡是人们犯了罪过，较大的就被夺除寿命十二年，较小的就被减去一百天。这些过错按轻重大小，种类有几百种之多。想要求得长生的人，首先必须避免这些过失。*<br>

**是道则进，非道则退。不履邪径，不欺暗室。积德累功，慈心于物。忠孝友悌，正己化人。矜[guān]孤恤寡，敬老怀幼。昆虫草木，犹不可伤。宜悯人之凶，乐人之善。**<br>
*合乎道理的就要前进去做，不合乎道理的就必须退避不为。不走邪恶之途，不在暗中做亏心事。积累德行与功业，对万物常怀慈悲之心。忠诚、孝顺，兄长友爱弟弟，弟弟尊敬兄长。端正自己并感化他人。怜悯孤独者，周济鳏寡者，尊敬衰老者，关怀年幼者。昆虫和草木，仍然不可以伤害。应该同情他人的不幸，欣赏他人的善行。*<br>
**济人之急，救人之危。见人之得，如己之得。见人之失，如己之失。不彰人短，不炫己长。遏恶扬善，推多取少。受辱不怨，受宠若惊。施恩不求报，与人不追悔。**<br>
*帮助他人的急难，救助他人的危困。看见别人有所得，就如同自己得到一样高兴；看见他人有所损失，就如同自己损失一样难过。不要彰扬别人的短处，不要炫耀自己的长处。阻止恶行，赞扬善行。多让给别人，少为自己取。受到侮辱也不怨恨。得到过分的宠爱待遇，感到意外，而不知所措。布施恩惠，不求回报；赠送给他人后，不再追溯后悔。*<br>

## 读后感
&emsp;&emsp;选择性吸收，完全按照这样做，太委屈自己了吧。遏恶扬善可以，推多取少就算了吧，除非在一个高度文明的社会。感觉描述地像一个共产主义社会，人人素质极高，物质精神极大丰富。
